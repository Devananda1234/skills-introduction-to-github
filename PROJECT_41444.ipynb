```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
travel_df=pd.read_csv('Travel info original.csv')
 #Visualization 1: Gender Distribution
plt.figure(figsize=(7, 5))
sns.countplot(data=travel_df, x='Gender', palette='coolwarm')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Visualization 2: Travel Frequency
plt.figure(figsize=(6,5))
sns.countplot(data=travel_df, x='Travel_Frequency', order=travel_df['Travel_Frequency'].value_counts().index, palette='mako')
plt.title('Travel Frequency Distribution')
plt.xlabel('Travel Frequency')
plt.ylabel('Count')
plt.show()

# Visualization 3: Age Distribution
plt.figure(figsize=(6,5))
sns.histplot(travel_df['Age'], color='purple')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()

# Visualization 4: Top Preferred Destinations
plt.figure(figsize=(6,5))
destinations = travel_df['Preferred_Destinations'].str.split(';').explode().str.strip()
top_destinations = destinations.value_counts().head(10)
sns.barplot(x=top_destinations.values, y=top_destinations.index, palette='flare')
plt.title('Top 10 Preferred Destinations')
plt.xlabel('Count')
plt.ylabel('Destination')
plt.show()

# Visualization 5: Pie Chart for Travel Style
travel_style_counts = travel_df['Travel_Style'].value_counts()
plt.figure(figsize=(8,5))
plt.pie(travel_style_counts, labels=travel_style_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Travel Style Distribution')
plt.show()

# Visualization 6 : Budget Distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=travel_df, y='Budget', order=travel_df['Budget'].value_counts().index, palette='Spectral')
plt.title('Budget Preferences')
plt.xlabel('Count')
plt.ylabel('Budget Range')
plt.show()
```
